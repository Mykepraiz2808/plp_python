```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set(style="whitegrid")
```

```python
# Step 2: Load the metadata.csv file
df = pd.read_csv("metadata.csv")

# Step 3: Inspect first rows
df.head()
```

```python
# Step 4: Basic exploration
print("Shape (rows, columns):", df.shape)
print("\nInfo:")
print(df.info())
```

```python
# Check missing values
print("\nMissing values:")
print(df.isnull().sum().head(20))  # show first 20 cols
```

```python
# Step 5: Basic statistics
df.describe()
```

```python
# Part 2: Data Cleaning and Preparation
# Step 6: Handle missing data
# Example: Drop rows without 'title' or 'publish_time'
df_clean = df.dropna(subset=["title", "publish_time"]).copy()
print("After cleaning:", df_clean.shape)
```

```python
# Step 7: Convert publish_time to datetime
df_clean["publish_time"] = pd.to_datetime(df_clean["publish_time"], errors="coerce")
# Extract year for analysis
df_clean["year"] = df_clean["publish_time"].dt.year

# Drop rows without a valid year
df_clean = df_clean.dropna(subset=["year"])
```

```python
# Step 8: Add abstract word count
df_clean["abstract_word_count"] = df_clean["abstract"].fillna("").apply(lambda x: len(x.split()))
df_clean[["title", "abstract_word_count"]].head()
```

```python
# Step 9: Count papers by publication year
year_counts = df_clean["year"].value_counts().sort_index()
print(year_counts)
```

```python
# Step 10: Top journals
top_journals = df_clean["journal"].value_counts().head(10)
print(top_journals)
```

```python
# Step 11: Frequent words in titles (basic)
from collections import Counter
import re

def get_title_words(titles):
    words = []
    for t in titles.dropna():
        words.extend(re.findall(r"\b\w+\b", t.lower()))
    return words

title_words = get_title_words(df_clean["title"])
common_words = Counter(title_words).most_common(20)
common_words
```

```python
# Plot publications over time
plt.figure(figsize=(10,5))
sns.barplot(x=year_counts.index, y=year_counts.values, color="skyblue")
plt.title("Publications by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# Bar chart of top journals
plt.figure(figsize=(10,5))
sns.barplot(y=top_journals.index, x=top_journals.values, color="salmon")
plt.title("Top Journals Publishing COVID-19 Research")
plt.xlabel("Count")
plt.ylabel("Journal")
plt.show()
```

```python
# Word cloud of titles
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(title_words))

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Paper Titles")
plt.show()
```

```python
# Distribution by source_x (dataset source)
plt.figure(figsize=(10,5))
df_clean["source_x"].value_counts().head(10).plot(kind="bar", color="lightgreen")
plt.title("Distribution of Paper Counts by Source")
plt.xlabel("Source")
plt.ylabel("Count")
plt.show()
```

```python
### Findings
- Publications increased sharply in 2020 due to COVID-19.
- Certain journals dominated the early research landscape.
- Word cloud shows frequent terms like "covid", "coronavirus", "pandemic".

### Challenges
- Handling missing and inconsistent publication dates.
- Dataset is very large; requires filtering and careful memory use.

### Learning
- Practiced pandas cleaning workflows.
- Learned basic visualization and Streamlit app deployment.
- Gained confidence in structuring a small data science project.
```
